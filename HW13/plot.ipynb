```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取CSV檔案
data = pd.read_csv('./pruning_results.csv')

# 提取資料
ratio = data['ratio']
accs = data['accs']

# 創建折線圖
plt.plot(ratio, accs, marker='o')

# 添加標題和軸標籤
plt.title('Pruning Ratio v.s. Model Accuracy')
plt.xlabel('Pruning Ratio')
plt.ylabel('Model Accuracy')

plt.xticks([i/20 for i in range(0, 20)], rotation=45) 
plt.yticks([i/10 for i in range(0, 11)])  

# 顯示圖表
plt.show()
```
